## Asymptotic Analysis: Big O and Other Notations

### What is Asymptotic Analysis?
Asymptotic analysis is a technique used to describe the behavior of an algorithm as the input size grows. It provides a way to evaluate the efficiency of an algorithm by focusing on its performance in terms of time and space complexity.

### Asymptotic notations

1. (Big–O, O()): Let f(n) and g(n) be functions that map positive integers to positive real
numbers. We say that f(n) is O(g(n)) (or f(n) ∈ O(g(n))) if there exists a real constant c > 0 and there
exists an integer constant n0 ≥ 1 such that f(n) ≤ c ∗ g(n) for every integer n ≥ n0. This means that for sufficiently large  n, the function f(n) will not grow faster than a constant multiple of g(n).

2. (Little–o, o()): Let f(n) and g(n) be functions that map positive integers to positive real
numbers. We say that f(n) is o(g(n)) (or f(n) ∈ o(g(n))) if for any real constant c > 0, there exists an
integer constant n0 ≥ 1 such that f(n) < c ∗ g(n) for every integer n ≥ n0.

3. n (Big–Omega, Ω()): Let f(n) and g(n) be functions that map positive integers to positive
real numbers. We say that f(n) is Ω(g(n)) (or f(n) ∈ Ω(g(n))) if there exists a real constant c > 0 and
there exists an integer constant n0 ≥ 1 such that f(n) ≥ c · g(n) for every integer n ≥ n0.

4. (Little–Omega, ω()): Let f(n) and g(n) be functions that map positive integers to positive
real numbers. We say that f(n) is ω(g(n)) (or f(n) ∈ ω(g(n))) if for any real constant c > 0, there exists
an integer constant n0 ≥ 1 such that f(n) > c · g(n) for every integer n ≥ n0.

Big O or Big Oh notation describes the upper bound of an algorithm's running time. It provides a worst-case scenario, ensuring that the algorithm will not run longer than a certain time. It is the most commonly used notation in algorithm analysis.

The concept of Big O notation was introduced by German mathematician Paul Bachmann in the late 19th century and later popularized by Edmund Landau. It became a fundamental part of algorithm analysis in computer science through the work of Donald Knuth and others in the mid-20th century. 

Here's a brief history of its development:
1. Paul Bachmann: Introduced the notation in the context of number theory.
2. Edmund Landau: Expanded and formalized the use of Big O notation in mathematics, especially in analytic number theory.
3. Donald Knuth: Brought Big O notation into computer science, making it a standard tool for analyzing algorithm efficiency.

The primary purpose of Big O notation is to provide a high-level understanding of the algorithm's performance and scalability. By focusing on the most significant term and ignoring constant factors and lower-order terms, Big O notation allows for a simplified comparison of different algorithms.).

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme()
````

````python
n = np.linspace(0.01, 10, 400)

f_n = 3 * n**2 + 2 * n + 1
g_n = n**2
c_g_n = 4 * n**2  # here c = 4

n0 = 2

plt.plot(n, f_n, label = "f(n)")
plt.plot(n, g_n, label = "g(n)")
plt.plot(n, c_g_n, label = "c.g(n)")
plt.axvline(x = n0, label = "n0", linestyle = "--")
plt.legend()
plt.xlabel('n')
plt.ylabel('Value')
plt.title('Big O Notation Visualization')
plt.show()
````

In algorithm analysis, several functions and notations are commonly used to describe the performance characteristics of algorithms. Here are some of the key ones:

* **Constant Time (O(1))**: An algorithm that always takes the same amount of time or space to execute, regardless of the size of the input. For example, accessing an element in an array by index or performing basic arithmetic operations. 

* **Linear Time (O(n))**: An algorithm whose performance grows linearly with the size of the input. For example, iterating through an array or list once.

* **Logarithmic Time (O(log n))**: An algorithm whose performance grows logarithmically with the size of the input. For example, binary search in a sorted array.

* **Quadratic Time (O(n^2))**: Think of a process that takes time proportional to the square of the input size. Nested loops.

* **Exponential Time (O(2^n))**: Time it takes to solve problem grows exponentially grows with size of n. 

* **Factorial Time (O(n!))**:  Algorithms with factorial time complexity have runtimes that grow factorially with the input size. Examples include brute-force algorithms that enumerate all permutations or combinations of a problem.

* **Polynomial Time (O(n^k))**: Algorithms with polynomial time complexity have runtimes that are bounded by a polynomial function of the input size. This includes quadratic (n^2), cubic (n^3), and higher-degree polynomials.

In computational complexity theory, the focus is on how algorithms behave as the input size approaches infinity. Asymptotic analysis, allows us to compare algorithms based on their growth rates.

Consider two algorithms, A and B, where:

- Algorithm A has a time complexity of O(n^2), but it has a small constant factor.
- Algorithm B has a time complexity of O(n log n), but it has a larger constant factor.

In practical terms, this means that for small input sizes, Algorithm A might outperform Algorithm B due to its smaller constant factor. However, as the input size grows larger and approaches infinity, Algorithm B will eventually become faster due to its better asymptotic complexity. Even though Algorithm B may have a larger overhead for smaller inputs, its superior scalability ensures that it will eventually surpass the performance of Algorithm A for sufficiently large inputs. This principle is crucial in real-world scenarios, where dealing with large datasets or inputs is common.

Notice in the graph above that exponential had better performance than cubic before 10 or n < 10 but after that it exploded!!

#### How to use Big-Oh in practice for your code

Identify the Basic Operations: Start by identifying the fundamental operations in your code. These are the operations that are repeated as the input size increases. Common basic operations include assignments, comparisons, arithmetic operations, function calls, loops, and recursion.

Determine the Dominant Term: Analyze which part of the code contributes the most to the overall runtime or space complexity. Focus on the part of the code that grows the fastest with increasing input size. This often involves identifying loops, recursive calls, or nested operations.

Count the Operations: Estimate how many times the identified basic operations are executed as a function of the input size. For loops, consider how many iterations occur based on the input size. For recursive algorithms, analyze the number of recursive calls made.

Simplify and Express as a Function of Input Size: Express the number of operations as a function of the input size, typically denoted as "n." Simplify the expression by removing lower-order terms, constants, and coefficients. This step helps focus on the growth rate rather than specific details.

Determine the Big O Notation: Once you have the simplified expression, determine its growth rate. The dominant term in the expression represents the complexity class, and you can express it using Big O notation. Remember that Big O notation describes an upper bound on the growth rate, so focus on the term that grows the fastest as "n" approaches infinity.

```Python
def example_function(arr):
    n = len(arr)
    for i in range(n):          # O(n)
        print(arr[i])           # O(1)
    
    for i in range(n):          # O(n)
        for j in range(n):      # O(n)
            print(arr[i] + arr[j])  # O(1)

    for i in range(n):          # O(n)
        for j in range(n):      # O(n)
            for k in range(n):  # O(n)
                print(arr[i] + arr[j] + arr[k])  # O(1)
```

In this code:

The first loop runs "n" times, contributing O(n).
The second loop runs "n * n" times, contributing O(n^2).
The third loop runs "n * n * n" times, contributing O(n^3).
Since the third loop dominates the runtime complexity, the overall complexity of this code is O(n^3).
